```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# 1)a. Start off by importing the cars.csv file in the jupyter notebook.

```python
cars = pd.read_csv('cars.csv')
cars
```

# 1)b. Generate a line plot graph for the column ‘model’ and ‘hp’.

```python
plt.plot( cars['model'],  cars['hp'])
plt.xlabel('Models')
plt.ylabel('hp')
plt.title ('Model Names vs. HorsePower')

plt.show()
```

```python
cars.columns
```

# 2) Generate a bar plot graph for the columns ‘carbs’ and ‘gear’.


```python
plt.bar(cars.carb , cars.gear )
plt.xlabel('Carbs')
plt.ylabel('gear')
plt.title('carbs vs. gear')
fig = plt.figure(figsize =(100, 709))
plt.show()
```

# 3) Plot a histogram for the column ‘wt’.


```python
plt.hist(cars.wt, bins = 30)
plt.xlabel('weight of the cars')
plt.ylabel('Count')
plt.title('Histogram for the Weight values')
plt.show()
```

# 4)  Plot a pie chart for columns: ‘cyl’ and ‘model’ form the mtcars.csv data frame

```python
plt.pie(cars.cyl , labels = cars.model, autopct= '%1.1f%%')
plt.show()
```

# 5) Plot the area chart for the colmns: ‘am’ and ‘carb’.

```python
plt.stackplot(cars.am , cars.carb)
plt.xlabel('Transmission')
plt.ylabel('Number Of Carburetors')
plt.title('Transmissions vs Number Of Carburetors')
plt.grid()
plt.show()
```
